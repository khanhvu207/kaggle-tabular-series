```python
import pandas as pd
```

```python
l1_oof = pd.read_parquet("data/l1_oof_preds.parquet")
l1_test = pd.read_parquet("data/l1_test_preds.parquet")
```

```python
train_df = pd.read_parquet("data/train_v3.parquet")
test_df = pd.read_parquet("data/test_v3.parquet")
```

```python
train_df["kfold"]
```

```python
orig_train_df = train_df.copy()
```

```python
orig_train_df[orig_train_df["kfold"] == 0]
```

```python
for fold_nr in range(5):
    indicies = orig_train_df[orig_train_df["kfold"] == fold_nr].index
    num_row = len(indicies)
    fold_ids = [i for i in range(5) if i != fold_nr]
    new_fold_ids = [fold_ids[i % 4] for i in range(num_row)]
    train_df.loc[indicies, "kfold"] = new_fold_ids
```

```python
train_df[train_df["kfold"] == 0]
```

```python
train_df.merge(l1_oof, on="id", how="left").to_parquet("data/l2_train.parquet")
test_df.merge(l1_test, on="id", how="left").to_parquet("data/l2_test.parquet")
```
